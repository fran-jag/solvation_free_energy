```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fnn_opt = pd.read_csv('../data/optimization_results/FFNN_optimization.csv')
gcn_opt = pd.read_csv('../data/optimization_results/GCN_optimization.csv')
rfr_opt = pd.read_csv('../data/optimization_results/RFR_optimization.csv')
xgb_opt = pd.read_csv('../data/optimization_results/XGB_optimization.csv')
```

```python
fnn_opt_best_rmse = fnn_opt['rmse'].cummin()
gcn_opt_best_rmse = gcn_opt['rmse'].cummin()
rfr_opt_best_rmse = rfr_opt['rmse'].cummin()
xgb_opt_best_rmse = xgb_opt['rmse'].cummin()
```

```python
opt_list = [fnn_opt_best_rmse, gcn_opt_best_rmse, xgb_opt_best_rmse, rfr_opt_best_rmse]
model_names = ['Feedforward Neural Network', 'Graph Convolutional Network', 'XGBoost Regressor', 'Random Forest Regressor']
plt.figure(dpi=150)
for i, line in enumerate(opt_list):
    plt.plot(range(21), line[:21], label=model_names[i])

plt.grid(True, alpha=0.3)
plt.yticks(np.arange(0, 4.55, 0.25))
plt.xticks(np.arange(0, 21, 5))
plt.xlim(0, 20)

plt.xlabel('Iteration')
plt.ylabel('RMSE')
plt.title('Best objective found vs. # of iterations')
plt.legend()
plt.show()

```
